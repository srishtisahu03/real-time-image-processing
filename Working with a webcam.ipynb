```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Working with a Webcam

```python
cam = cv.VideoCapture(0)

while True:

    _, img = cam.read()
    cv.imshow("Frame",img)

    key = cv.waitKey(30) #bigger the number, less smoother the frame (1 is the smoothest)

    if (key==27):
        cam.release()
        break
```

### Flipping

```python
cam = cv.VideoCapture(0)

while True:

    _, img = cam.read()

    flip = cv.flip(img, 1)

    cv.imshow("Frame",img)
    cv.imshow("Flipped",flip)

    key = cv.waitKey(1) #bigger the number, less smoother the frame (1 is the smoothest)

    if (key==27):
        cam.release()
        break
```

### Cropping

```python
cam = cv.VideoCapture(0)

while True:

    _, img = cam.read()
    img = cv.flip(img, 1)

    frame = img[100:500, : , : ]

    cv.imshow("Frame",img)
    cv.imshow("Cropped",frame)

    key = cv.waitKey(1) #bigger the number, less smoother the frame (1 is the smoothest)

    if (key==27):
        cam.release()
        break
```

### Frame extraction

```python
cam = cv.VideoCapture(0)

while True:

    _, img = cam.read()

    img = cv.flip(img, 1)
    hsv = cv.cvtColor(img, cv.COLOR_RGB2HSV)

    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
     
    h = hsv[:,:,0]
    s = hsv[:,:,1]
    v = hsv[:,:,2]

    cv.imshow("Frame", img)
    cv.imshow("Red", r)
    cv.imshow("Green", g)
    cv.imshow("Blue", b)
    
    cv.imshow("HSV", hsv)
    cv.imshow("Hue", h)
    cv.imshow("Saturation", s)
    cv.imshow("Value", v)

    if (key == 27):
        cam.release()
        break
```
